```python
#now we will scrape for EVERY team and do it across multiple seasons to be used in our training of our model later
years = list(range(2025, 2020,-1))
years
```

```python
all_matches = []
```

```python
all_matches
```

```python
standings_url = "https://fbref.com/en/comps/20/Bundesliga-Stats"
```

```python
standings_url
```

```python
for year in years:
    #standings_url must be updated to form of: https://fbref.com/en/comps/9/2023-2024/2023-2024-Bundesliga-Stats each 
    #iteration to be on the right page
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text)
    standings_table = soup.select('table.stats_table')[0]

    links = standings_table.find_all('a')
    links = [l.get("href") for l in links]
    links = [l for l in links if '/squads/' in l]
    #the season year (2023-2024) should updating automatically so long as standings_url is updated correctly
    team_urls = [f"https://fbref.com{l}" for l in links]

    previous_season = soup.select("a.button2.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"

    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-"," ")
        
        data = requests.get(team_url)
        soup = BeautifulSoup(data.text)
        scores_table = soup.find("table", id="matchlogs_for")
        html_version_scores = str(scores_table)
        matches = pd.read_html(StringIO(html_version_scores))[0] # already taking first element in list no need to do it in next line

        # no need to create another BeautifulSoup instance as we are on the same HTML page
        links = [l.get("href") for l in soup.find_all("a")]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        data = requests.get(f"https://fbref.com{links[0]}")
        #new instance as we are now on the shooting stats page
        soup = BeautifulSoup(data.text)
        shooting_table = soup.find("table", id="matchlogs_for")
        html_version_shooting = str(shooting_table)
        shooting = pd.read_html(StringIO(html_version_shooting))[0]
        shooting.columns = shooting.columns.droplevel()

        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError:
            continue

        team_data = team_data[team_data["Comp"] == "Bundesliga"]
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        time.sleep(6)
```

```python
match_df = pd.concat(all_matches)
```

```python
match_df.columns = [c.lower() for c in match_df.columns]
```
