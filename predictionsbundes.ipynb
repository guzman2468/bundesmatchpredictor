```python
matches["date"] = pd.to_datetime(matches["date"]) #ML can only work with numerical data so convert the date string to a date/time data type (similar to SQL)
matches["venue_code"] = matches["venue"].astype("category").cat.codes #convert the home/away into a category then into number (0 for away 1 for home)
matches["opp_code"] = matches["opponent"].astype("category").cat.codes # same as for venue but assigning a opponent team a number
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int") # remove the colon and minutes and keep the hour as an integer
matches["day_code"] = matches["date"].dt.dayofweek # take the date and get the day of the week a game was played on and assign it a integer
matches["target"] = matches["result"].map({"L": 0, "D": 1, "W": 2})  # set each match result to a number so our model can understand 
```

```python
matches #ensuring that everything worked
```

```python
#now we can start training the model
from sklearn.ensemble import RandomForestClassifier #non-linear model that knows that for ex: a teams opponent code of 
#18 doesn't mean that has a relationship to how good the team is
# easy to use for this case
```

```python
rf = RandomForestClassifier(n_estimators=100, min_samples_split=50, random_state=1)
#make it so we have 75 decision trees and make it so the each node in the tree doesn't split unless we have AT LEAST 15 samples, and to ensure our random data is not completely "random" each time set a random state 
```

```python
train = matches[matches["date"] < '2023-01-01']  #roughly having 2.5 seasons worth of training data and 2 seasons of test data
test = matches[matches["date"] > '2023-01-01'] #split the data (roughly) in half to be able to compare our test data to the predictions made by the training data
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"] #to predict, we will be using 1. Home or Away, 2. What opponent it is 3. The time of day and 4. What day of the week it is
```

```python
rf.fit(train[predictors], train["target"]) #training the RF model using the predictors to see if it can match our target(the actual result of the match)
#potentially add more predictors later to increase accuracy
```

```python
preds = rf.predict(test[predictors]) #selecting the cols stated in predictors from the TEST data and then we are applying the model to the data in the test set
```

```python
# how will we determine the accuracy of the model? by using a variety different metrics
from sklearn.metrics import accuracy_score #tells us what percent of time our prediction was accurate
```

```python
acc = accuracy_score(test["target"], preds) #comparing how accurate our preds were to our actual "target" values in the test data set
```

```python
acc
# the value score is not great, but this is partially due to the fact that we created 3 different outcomes, 0 for loss, 1 for draw, and 2 for win. This means
# that our model has more room for error in the prediction but can give us a clearer idea on what the outcome would be
# if we wanted this number to be better, we could generalized loss and draw to 0 and 1 as a win
# all in all, I gave up the ability to predict WINS better for the ability to be able to distinguish whether a team lost or drew a game
```

```python
combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
pd.crosstab(index=combined["actual"], columns=combined["prediction"])

# Analysis: When a loss was predicted, we were right around 47% of the time (not terrible but it could be better)
# When a draw was predicted, we were right around 25% of the time (not very good at all)
# When a win was predicted, we were right around 46% of the time (not terrible)

# In general, it can be seen that the model was correct AROUND 50% of the time when predicting a result.
# Again, I sacrificed the ability to guess wins correct more often in exchange to differentiate between when a loss or a draw happend
# This could potentially be more helpful when trying to more accurately define what the result of a match may be
```

```python
#improve accuracy using rolling average so that we can focus more on long-term trends in our results and not short term changes(i.e caring more about a how a team has done across the ENTIRE season as opposed to potentially
# having a bad run of form
grouped_matches = matches.groupby("team") # 1 df per team
```

```python
group = grouped_matches.get_group("Bayern Munich")
```

```python
group.head()
```

```python
def rolling_averages(group, cols, new_cols):
    #parameters are the team, columns we want to compute rolling avg for and new_cols that we assign the rolling avgs to
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean() #previous 3 weeks and closed=left to make sure the current week is left out when creating rolling avg
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols) #removes rows w/ missing values, i.e week 2 because there is no prev 3 weeks to look at
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols] # we will add these new rolling ___ columns to improve accuracy of predictions
```

```python
rolling_averages(group, cols, new_cols) #starts at week 4
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
#create a df for each team then apply our rolling_averages function on each df
# feature may not work in future pandas versions
```

```python
def make_predictions(data, predictors):
    # using this function so that code doesn't need to be typed out line by line each time we want to make a new prediction(i.e making changes in our data)
    train = data[data["date"] < '2023-01-01']
    test = data[data["date"] > '2023-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds, average="weighted") # again using weighted to account for imbalances in our classes
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols) #predictors is OG set of 4 columns we used
```
